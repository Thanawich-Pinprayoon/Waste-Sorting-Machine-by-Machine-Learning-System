```
import os
from google.colab import drive
drive.mount('/content/drive/')
```

#Change Name

```
filename_list = os.listdir('/content/drive/My Drive/CE_Project/Dataset/Book')
name = 'book'
for filename_index in range(len(filename_list)):
  if '.' in filename_list[filename_index]:
    src = '/content/drive/My Drive/CE_Project/Dataset/Book' + '/' + filename_list[filename_index]
    dst = '/content/drive/My Drive/CE_Project/Dataset/Book' + '/' + str(name)+str(filename_index)
    os.rename(src,dst)
```
