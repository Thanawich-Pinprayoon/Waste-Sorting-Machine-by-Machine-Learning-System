#Change Name All File

```
dataset_type_list = os.listdir('/content/drive/My Drive/CE_Project/Dataset/') #get all type name

print('found dataset ' + str(len(dataset_type_list)) + ' type')


for dataset_type in dataset_type_list:
  print(' ' + str(dataset_type))
  dataset_items_list = os.listdir('/content/drive/My Drive/CE_Project/Dataset/' + dataset_type ) #get all items name
  print('   found item ' + str(len(dataset_items_list)) + ' items')

  for dataset_item in dataset_items_list:
    print('     ' + str(dataset_item))
    filename_list = os.listdir('/content/drive/My Drive/CE_Project/Dataset/' + str(dataset_type) + '/' + str(dataset_item) ) #get all item filename list

    print('         found picture ' + str(len(filename_list)) + ' pics in ' + str(dataset_item) + ' from ' + str(dataset_type))

    name = str(dataset_item)

    for filename_index in range(len(filename_list)):
      if '.' in filename_list[filename_index]:
        src = '/content/drive/My Drive/CE_Project/Dataset/' + str(dataset_type) + '/' + str(dataset_item) + '/' + filename_list[filename_index]
        dst = '/content/drive/My Drive/CE_Project/Dataset/' + str(dataset_type) + '/' + str(dataset_item) + '/' + str(name)+str(filename_index)
        os.rename(src,dst)
```
