```python
import os
import sys
import shutil
```

```python
dir = '/content/gdrive/MyDrive/CE_Project/Dataset/labeled/images/'
for f in os.listdir(dir):
    os.remove(os.path.join(dir, f))
dir = '/content/gdrive/MyDrive/CE_Project/Dataset/labeled/labels/'
for f in os.listdir(dir):
    os.remove(os.path.join(dir, f))
```

```python
types = [
         'Hazardous waste',
         'General',
         'Recycle'
         ]
```

```python
sfor i in types:
  dir = '/content/gdrive/MyDrive/CE_Project/Dataset/resized/'+i
  os.chdir(dir)
  fileNames_img = os.listdir()

  dir = '/content/gdrive/MyDrive/CE_Project/Dataset/resized/'+i+'/labeled'
  os.chdir(dir)
  fileNames_txt = [os.path.splitext(filename)[0] for filename in os.listdir()]
  print(str(len(fileNames_txt)-1)+' images from '+i+' has been labeled.')

  for fileName in fileNames_txt:
    if 'label' in fileName.lower() or 'desktop' in fileName.lower():
      continue
    else:
      for imgfile in fileNames_img:
        x = imgfile.split(".")
        if fileName == x[0]:
          print('     '+fileName)
          print('     '+imgfile+' found! moving it')
          shutil.copyfile('/content/gdrive/MyDrive/CE_Project/Dataset/resized/'+i+'/'+imgfile,
                          '/content/gdrive/MyDrive/CE_Project/Dataset/labeled/images/'+imgfile)
          
          shutil.copyfile('/content/gdrive/MyDrive/CE_Project/Dataset/resized/'+i+'/labeled'+'/'+fileName+'.txt',
                          '/content/gdrive/MyDrive/CE_Project/Dataset/labeled/labels/'+fileName+'.txt')
          break
  dir = '/content/gdrive/MyDrive/CE_Project/Dataset/labeled/images'
  imgs = os.listdir(dir)
  print(' '+str(len(imgs))+' images from '+i+' moved to labeled.')
  dir = '/content/gdrive/MyDrive/CE_Project/Dataset/labeled/labels'
  lbls = os.listdir(dir)
  print(' '+str(len(imgs))+' label files from '+i+' moved to labeled.')
```
