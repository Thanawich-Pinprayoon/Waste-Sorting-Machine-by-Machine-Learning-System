```python
import pylab as plt
```

```python
fig, ax = plt.subplots()
result_plot = sns.barplot(data=result_pd, ax=ax, x="Class", y="Count")

def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .35)
# result_plot.get_legend().remove()
plt.setp(result_plot.get_xticklabels(), rotation=90)
# plt.ylim(0, 1200)
result_plot.tick_params(labelsize=14)
result_plot.tick_params(length=0, axis='x')
result_plot.set_ylabel("Count", fontsize=15)
# result_plot.set_xlabel("", fontsize=1)
plt.tight_layout()
plt.show()
```
