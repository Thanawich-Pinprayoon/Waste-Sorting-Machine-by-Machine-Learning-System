```python
for fileName in fileNames:
  for aug in range(5):
    for k in classlist:
      if k in fileName:
        try:
          shutil.move("/content/gdrive/MyDrive/CE_Project/Dataset/Augmented/images/aug" + str(aug) +"/" + fileName+ "_aug" + str(aug) + ".jpg",
                         "/content/gdrive/MyDrive/CE_Project/Dataset/Augmented/images/aug" + str(aug) +"/" + k + "/" + fileName+ "_aug" + str(aug) + ".jpg")
        except:
          print("/Augmented/images/aug" + str(aug) +"/" + fileName+ "_aug" + str(aug) + ".jpg not found")
        try:
          shutil.move("/content/gdrive/MyDrive/CE_Project/Dataset/Augmented/labels/aug" + str(aug) +"/" + fileName+ "_aug" + str(aug) + ".txt",
                         "/content/gdrive/MyDrive/CE_Project/Dataset/Augmented/labels/aug" + str(aug) +"/" + k + "/" + fileName+ "_aug" + str(aug) + ".txt")
        except:
          print("/Augmented/labels/aug" + str(aug) +"/" + fileName+ "_aug" + str(aug) + ".txt not found")
```
