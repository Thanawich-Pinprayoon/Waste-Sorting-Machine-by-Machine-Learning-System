```python
import albumentations as A
import cv2
import os
import sys
```

```python
 classlist = ["Correction_fluid","Battery","Cleaning_solution","Light_bulb","phone","Insecticide","Three_eye_plug","Aerosol_cans","Snack","Tissue","Foam_box","Paper_cup","Plastic_box","Drinking_straw","paper","plastic_glass","bottle","can","carton","glass_bottle","news_paper","plastic_bag","plastic_bottle"]
```

```python
fileNames = [os.path.splitext(filename)[0] for filename in os.listdir("/content/gdrive/MyDrive/CE_Project/Dataset/labeled/labels/")]
```

```python
fileNames.sort()
```

```python
for fileName in fileNames:
  if not "plastic_glass" in fileName:
    continue
  print(fileName)
  if '.ini' in fileName:
    continue
  for aug in range(5):
    image = cv2.imread("/content/gdrive/MyDrive/CE_Project/Dataset/labeled/images/"+fileName+".jpg")
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    with open("/content/gdrive/MyDrive/CE_Project/Dataset/labeled/labels/"+fileName+".txt") as file:
      bboxes = []
      for line in file:
        # print(line.rstrip())
        bbox = line.rstrip()
        bbox_class = bbox.split(" ")
        # print(bbox_class)
        bbox_class.append(int(bbox_class.pop(0)))
        # bbox_class.pop()
        # print(bbox_class)
        for i in range(len(bbox_class)-1):
          bbox_class[i] = float(bbox_class[i])
          if bbox_class[i] <= 0:
            bbox_class[i] = 0.000001
        # print(bbox_class)
        
        bboxes.append(bbox_class)
        # print("")
        # print("")
        # print(bboxes)

    if aug == 0:
      transform = A.Compose([
        # A.RandomSizedCrop(p=0.5,w2h_ratio = 1.0, height=480, width= 480,min_max_height = [240,240]),
        A.HorizontalFlip(p=1),
        A.RandomBrightnessContrast(p=0.2),
        A.Blur(p=0.2)
      ], bbox_params=A.BboxParams(format='yolo', min_visibility=0.9))
    elif aug == 1:
      transform = A.Compose([ 
        # A.RandomSizedCrop(p=0.5,w2h_ratio = 1.0, height=480, width= 480,min_max_height = [240,240]),
        A.VerticalFlip(p=1),
        A.RandomBrightnessContrast(p=0.2),
        A.Blur(p=0.2)
      ], bbox_params=A.BboxParams(format='yolo', min_visibility=0.9))
    elif aug == 2:
      transform = A.Compose([
        # A.RandomSizedCrop(p=0.5,w2h_ratio = 1.0, height=480, width= 480,min_max_height = [240,240]),
        A.HorizontalFlip(p=1),
        A.VerticalFlip(p=1),
        A.RandomBrightnessContrast(p=0.2),
        A.Blur(p=0.2)
      ], bbox_params=A.BboxParams(format='yolo', min_visibility=0.9))
    elif aug == 3:
      transform = A.Compose([
        # A.RandomSizedCrop(p=0.5,w2h_ratio = 1.0, height=480, width= 480,min_max_height = [240,240]),
        A.RandomBrightnessContrast(p=0.2),
        A.Cutout(p=1),
      ], bbox_params=A.BboxParams(format='yolo', min_visibility=0.9))
    elif aug == 4:
      transform = A.Compose([
        # A.RandomSizedCrop(p=0.5,w2h_ratio = 1.0, height=480, width= 480,min_max_height = [240,240]),
        A.RandomSunFlare(p=1),
        A.RandomBrightnessContrast(p=0.2),
        A.Blur(p=0.2)
      ], bbox_params=A.BboxParams(format='yolo', min_visibility=0.9))
    # print(bboxes)
    transformed = transform(image=image, bboxes=bboxes)
    # print(transformed)
    transformed_image = transformed['image']
    
    transformed_bboxes = transformed['bboxes']
    # print(transformed_bboxes)
    # sys.exit()

    # save txt
    result = ''.join([i for i in fileName if not i.isdigit()])
  
    for classs in classlist:    
      #print(result+" "+classs)
      if classs == result:
        # print("==============================================================")
        f= open("/content/gdrive/MyDrive/CE_Project/Dataset/Augmented/labels/aug"+str(aug)+"/"+result+"/"+fileName+"_aug"+str(aug)+".txt","w+")
        for box_index in range(len(transformed_bboxes)):
          label = list(transformed_bboxes[box_index])
          class_num = label.pop()
          label.insert(0, class_num)
          # print(label)
          aLine = ""
          for value in label:
            if value is int(value):
              aLine = aLine+str(value)
            else:
              if value is 1.0:
                aLine = aLine+"0.999999"
                print("============",temp,"  " ,value ,"========")
              else:
                aLine = aLine + "{:.6f}".format(value)
            aLine = aLine + " "
          f.write(aLine.strip())
          f.write("\n")
        f.close()
        # save img
        cv2.imwrite("/content/gdrive/MyDrive/CE_Project/Dataset/Augmented/images/aug"+str(aug)+"/"+result+"/"+fileName+"_aug"+str(aug)+".jpg",transformed_image)
```
