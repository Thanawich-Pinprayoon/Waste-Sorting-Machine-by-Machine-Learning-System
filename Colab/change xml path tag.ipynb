```python
from xml.etree import ElementTree as et
```

```python
count_xmls = []
for folder in folders:
    os.chdir("D:/Downloads/Compressed/224/datasets/xml/"+folder)
    files = os.listdir()
    xmls = [file for file in files if file.endswith(('xml'))]
    xmls.sort()
    count_xmls.append(len(xmls))
    for i in range(len(xmls)):
        tree = et.parse(
            "D:/Downloads/Compressed/224/datasets/xml/"+folder+"/"+xmls[i])
        tree.find(
            './/filename').text = str(xmls[i].replace(".xml", ".jpg"))
        tree.find('.//path').text = str("/content/maix_train/out/datasets/datasets/images/" +
                                        folder+"/"+xmls[i].replace(".xml",".jpg"))

        
        tree.write("D:/Downloads/Compressed/224/datasets/xml/" +
                   folder+"/"+xmls[i])
```
