```python
from PIL import Image
%cd '/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space'
types = [
         'Hazardous waste',
        #  'General',
        #  'Recycle'
         ]
for i in types:
    dir = '/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space/'+i
    os.chdir(dir)
    fileNames = os.listdir()
    for fileName in fileNames:
      if 'label' in fileName.lower():
        continue
      elif 'jpg' in fileName.lower() or 'png' in fileName.lower():
        image = PIL.Image.open(fileName)
        image = image.convert('RGB')
        width, height = image.size 
        image_resized = image.resize((480,480), PIL.Image.ANTIALIAS)
        width, height = image_resized.size 
        if width == height:
          image_resized.save('/content/gdrive/MyDrive/CE_Project/Dataset/resized/'+i+'/'+fileName)
        else:
          print('     ->>>>>>>>'+fileName + ' is not squre')
```
