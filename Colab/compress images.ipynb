```python
from PIL import Image
import PIL
import os
import glob
```

```python
def compress_images(dir=False, quality=65):
    if dir:
        os.chdir(dir)
    files = os.listdir()
    images = [file for file in files if file.endswith(('jpg'))]
    for image in images:
        print(image)
        img = Image.open(image)
        img.save(image, optimize=True, quality=quality)
```

```python
compress_images("D:/Downloads/Compressed/no_aug/images", quality=10)
```
