```python
afrom PIL import Image
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space/Hazardous waste')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space/General')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space/Recycle')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/resized')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/resized/Hazardous waste')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/resized/General')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/resized/Recycle')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/need_crop/')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/need_crop/Hazardous waste')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/need_crop/General')
os.mkdir('/content/gdrive/MyDrive/CE_Project/Dataset/need_crop/Recycle')


%cd '/content/gdrive/MyDrive/CE_Project/Dataset'
types = ['Hazardous waste', 'General', 'Recycle']
for i in types:
  dir = '/content/gdrive/MyDrive/CE_Project/Dataset/'+i+'/'
  os.chdir(dir)
  classNames = os.listdir()
  for j in classNames:
    if j == 'labeled': 
      continue
    dir = '/content/gdrive/MyDrive/CE_Project/Dataset/'+i+'/'+j
    os.chdir(dir)
    fileNames = os.listdir()
    for fileName in fileNames:
      if 'label' in fileName.lower():
        continue
      elif 'jpg' in fileName.lower() or 'png' in fileName.lower():
        image = PIL.Image.open(fileName)
        image = image.convert('RGB')
        width, height = image.size 
        if width < height:
          new_width  = 480
          new_height = new_width * height // width
        else:
          new_height = 480
          new_width  = new_height * width // height
        image_resized = image.resize((new_width,new_height), PIL.Image.ANTIALIAS)
        if new_width == new_height:
          image_resized.save('/content/gdrive/MyDrive/CE_Project/Dataset/resized/'+i+'/'+fileName)
        else:
          image_resized.save('/content/gdrive/MyDrive/CE_Project/Dataset/need_crop/'+i+'/'+fileName)
          if new_width > new_height:
            result = Image.new(image_resized.mode, (new_width, new_width), (255, 255, 255))
            result.paste(image_resized)
          else:
            result = Image.new(image_resized.mode, (new_height, new_height), (255, 255, 255))
            result.paste(image_resized)S
          result.save('/content/gdrive/MyDrive/CE_Project/Dataset/added_white_space/'+i+'/'+fileName)
```
